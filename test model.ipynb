```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/home/virus/Downloads/sentdex_dog.png", cv2.IMREAD_GRAYSCALE)
img, test_img = cv2.resize(img, (256,256)), cv2.resize(img, (256,256))
plt.imshow(img, cmap=plt.cm.binary)
img = np.asarray([img])
img = np.expand_dims(img, -1)
img.shape
```

```python
model = tf.keras.models.load_model("model.model")
```

```python
predictions = model.predict(img)
prediction = np.argmax(predictions)
print(prediction)
```

```python
plt.imshow(test_img, cmap=plt.cm.binary)
if prediction == 0:
    print("catto")
elif prediction == 1:
    print("doggo")
```
